1. Определить корни


```python
from sympy import solveset
from sympy.abc import x, y
solveset(-12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30, x)
```

2. Найти интервалы, на которых функция возрастает

```python
from sympy import is_decreasing, is_increasing
from sympy import S, Interval, oo, sin, cos

is_increasing(-12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30, Interval.open(100, 100))
```

3. Найти интервалы, на которых функция убывает

```python
from sympy import is_decreasing, is_increasing
from sympy import S, Interval, oo, sin, cos

is_decreasing(-12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30, Interval.open(100, 100))
```

4. Построить график

```python
import sympy
from sympy import *
x = symbols('x')

plot(-12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30, (x, -100, 100))
```

5. Вычислить вершину

```python
# x = -b/2a - вершина функции
# x = 0

from sympy import Interval, Symbol, minimum, S
minimum(-12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30, x)
```
